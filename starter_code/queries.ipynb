# Answers
### 1. All the companies that it's name match 'Babelgum'. Retrieve only their `name` field.

```python
for i,x in enumerate(db.companies.find({'name':'Babelgum'}, {'name':1,'_id':0})):
    if i<50:
        print(x,'\n')
```

### 2. All the companies that have more than 5000 employees. Limit the search to 20 companies and sort them by **number of employees**.

```python
for i,x in enumerate(db.companies.find(filter={'number_of_employees':{'$gt':5000}},
                                       sort=[('number_of_employees',-1)],limit=(20),
                                       projection={'name':1,'_id':0,'number_of_employees':1})):
    if i<50:
        print(x,'\n')
```

### 3. All the companies founded between 2000 and 2005, both years included. Retrieve only the `name` and `founded_year` fileds.

### 15. Order the companies by their `founded year`, retrieving only their `name` and `founded year`.

```python
db.companies.create_index([("founded_year", DESCENDING)])
for i,x in enumerate(db.companies.find(filter={'founded_year':{'$not':{'$eq':None}}},
                                       sort=[('founded_year',1)],
                                       projection={'name':1,'_id':0,'founded_year':1})):
    if i<50:
        print(x,'\n')
```
